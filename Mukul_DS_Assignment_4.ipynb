```python
x = df[['Height']] # assign Height to x. we can also use x = df.drop('Weight',axix=1)instead
```

```python
y = df[['Weight']]   #assign Weight to y. we can use x = df.drop('Height',axix=1) instead
```

```python
#x.head()  #test output x after assigning 
```

```python
#y.head()  #test output y after assigning 
```

```python
from sklearn.linear_model import LinearRegression                      #importing Linear regression to apply .
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[60]])
```

```python
reg.predict(x)
```

```python
df['Predicted_Weight'] = reg.predict(x) #Create a new column with the name Predicted Weight
```

```python
plt.plot(df.Height,reg.predict(x)) #Showing best line here X axis is Height and Y axis is Weight
plt.scatter(df.Height,df.Weight)
plt.xlabel("Height")               
plt.ylabel("Weight")               
plt.title("Height vs Weight") 
```

```python
reg.score(x,y)  #checking the training accuracy
```

```python
from sklearn.metrics import mean_squared_error 
```

```python
mean_squared_error(df[['Height']], reg.predict(df[['Height']])) # counting error 
```

```python
mean_squared_error(x,df,Predicted_Weight)
```
